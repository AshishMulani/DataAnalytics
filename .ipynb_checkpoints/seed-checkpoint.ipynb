```python
import numpy as np
```

### Without seed

```python
# Without seed (different output every run)
print("Without seed (run 1) : ")
print("randint : ", np.random.randint(1,10,5)) # Generates 5 random integers between 1 and 10
print("rand : ", np.random.rand(5)) 
print("choice : ", np.random.choice([1,2,3,4],5))
```

### With Seed

```python
# With seed (same output every run)
np.random.seed(42)
print("With seed (run 1 ) : ")
print("randint : ", np.random.randint(1,10,5)) # Generates 5 random integers between 1 and 10
print("rand : ", np.random.rand(5)) 
print("choice : ", np.random.choice([1,2,3,4],5))
```

### uniform and normal

```python
print("\nuniform :", np.random.uniform(10,50,5))
print("normal :",np.random.normal(50,5,5))
```
