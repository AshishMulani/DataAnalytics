```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.

df = pd.read_csv("alphabet_stock_data.csv")

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

start_date = '2020-04-01'
end_date = '2020-09-24'

filtered_df = df.loc[start_date:end_date]

plt.figure(figsize=(10, 5))
plt.plot(filtered_df.index, filtered_df['Close'], marker='o', linestyle='-', color='b')

plt.title('Alphabet Inc. Historical Stock Prices', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Stock Price (USD)', fontsize=12)
plt.grid(True)

plt.show()
```

```python
#2. Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates.

start_date = '2020-04-01'
end_date = '2020-09-24'

filtered_df = df.loc[start_date:end_date]

plt.figure(figsize=(10, 6))
plt.plot(filtered_df.index, filtered_df['Open'], marker='o', label='Opening Price', color='green')
plt.plot(filtered_df.index, filtered_df['Close'], marker='o', label='Closing Price', color='red')

plt.title('Alphabet Inc. Open vs Close Prices', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend() 
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
#3. Write a Pandas program to create a bar plot of the trading volume of Alphabet Inc. stock between two specific dates.

start_date = '2020-04-01'
end_date = '2020-05-24'

filtered_df = df.loc[start_date:end_date]

plt.figure(figsize=(10, 6))
plt.bar(filtered_df.index.strftime('%Y-%m-%d'), filtered_df['Volume'], color='orange', alpha=0.8)

plt.title('Alphabet Inc. Trading Volume', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Volume (Number of Shares)', fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#4. Write a Pandas program to create a bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates

start_date = '2020-04-01'
end_date = '2020-05-24'

filtered_df = df.loc[start_date:end_date, ['Open', 'Close']]

filtered_df.index = filtered_df.index.strftime('%Y-%m-%d')

ax = filtered_df.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'salmon'], width=0.8)

plt.title('Alphabet Inc. Open vs Close Prices (Grouped Bar Chart)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Price Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#5. Write a Pandas program to create a stacked bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.
start_date = '2020-04-01'
end_date = '2020-05-24'

filtered_df = df.loc[start_date:end_date, ['Open', 'Close']]

filtered_df.index = filtered_df.index.strftime('%Y-%m-%d')

filtered_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#1f77b4', '#ff7f0e'])

plt.title('Alphabet Inc. Stacked Open & Close Prices', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Price Type')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#6. Write a Pandas program to create a horizontal stacked bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.

start_date = '2020-04-01'
end_date = '2020-05-24'

filtered_df = df.loc[start_date:end_date, ['Open', 'Close']]

filtered_df.index = filtered_df.index.strftime('%Y-%m-%d')

filtered_df.plot(kind='barh', stacked=True, figsize=(10, 6), color=['#4CAF50', '#FF5722'])

plt.title('Alphabet Inc. Horizontal Stacked Open & Close Prices', fontsize=14)
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Date', fontsize=12)
plt.legend(title='Price Type', loc='lower right')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#7. Write a Pandas program to create a histograms plot of opening, closing, high,low stock prices of Alphabet Inc. between two specific dates.
start_date = '2020-04-01'
end_date = '2020-05-24'

filtered_df = df.loc[start_date:end_date, ['Open', 'High', 'Low', 'Close']]

filtered_df.hist(bins=10, figsize=(12, 8), color='skyblue', edgecolor='black', grid=False)

plt.suptitle('Alphabet Inc. Price Distributions (Open, High, Low, Close)', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
#8. Write a Pandas program to create a stacked histograms plot of opening, closing, high, low stock prices of Alphabet Inc. between two specific dates.
filtered_df = df.loc[start_date:end_date, ['Open', 'Close', 'High', 'Low']]

filtered_df.plot(kind='hist', stacked=True, bins=20, figsize=(10, 6), edgecolor='black')
plt.title('Stacked Histograms of Alphabet Inc. Stock Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.legend(title='Price Type')
plt.show()
```

```python
#9. Write a Pandas program to draw a horizontal and cumulative histograms plot of opening stock prices of Alphabet Inc. between two specific dates.
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
filtered_df['Open'].plot(kind='hist', orientation='horizontal', color='skyblue', edgecolor='black')
plt.title('Horizontal Histogram (Opening Price)')
plt.xlabel('Frequency')
plt.ylabel('Price (USD)')

plt.subplot(1, 2, 2)
filtered_df['Open'].plot(kind='hist', cumulative=True, color='salmon', edgecolor='black')
plt.title('Cumulative Histogram (Opening Price)')
plt.xlabel('Price (USD)')
plt.ylabel('Cumulative Frequency')

plt.tight_layout()
plt.show()
```

```python
#12. Write a Pandas program to create a plot of stock price and trading volume of Alphabet Inc. between two specific dates.


start_date = '2020-04-01'
end_date = '2020-09-24'
filtered_df = df.loc[start_date:end_date]

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(filtered_df.index, filtered_df['Close'], color='blue', label='Stock Price')
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price (USD)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.bar(filtered_df.index, filtered_df['Volume'], color='gray', alpha=0.3, label='Volume')
ax2.set_ylabel('Trading Volume', color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

plt.title('Alphabet Inc. Stock Price and Trading Volume')
plt.show()
```

```python
#13. Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of Alphabet Inc. between two specific dates.

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True, gridspec_kw={'height_ratios': [3, 1]})

filtered_df[['Open', 'High', 'Low', 'Close', 'Adj Close']].plot(ax=ax1)
ax1.set_title('Alphabet Inc. Stock Prices and Volume')
ax1.set_ylabel('Price (USD)')
ax1.legend(loc='upper left')
ax1.grid(True, linestyle='--', alpha=0.7)

ax2.bar(filtered_df.index, filtered_df['Volume'], color='gray', alpha=0.5)
ax2.set_ylabel('Volume')
ax2.set_xlabel('Date')

plt.tight_layout()
plt.show()
```

```python
#14. Write a Pandas program to create a plot of adjusted closing prices, thirty days and forty days simple moving average of Alphabet Inc. between two specific dates.

df['30_SMA'] = df['Adj Close'].rolling(window=30).mean()
df['40_SMA'] = df['Adj Close'].rolling(window=40).mean()

start_date = '2020-04-01'
end_date = '2020-09-24'
filtered_df = df.loc[start_date:end_date]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df.index, filtered_df['Adj Close'], label='Adj Close', color='blue')
plt.plot(filtered_df.index, filtered_df['30_SMA'], label='30 Day SMA', color='red')
plt.plot(filtered_df.index, filtered_df['40_SMA'], label='40 Day SMA', color='green')

plt.title('Alphabet Inc. Adj Close with 30 & 40 Days SMA')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 15. Write a Pandas program to create a plot of adjusted closing prices, 30 days simple moving average and exponential moving average of Alphabet Inc. between two specific dates.

df['30_SMA'] = df['Adj Close'].rolling(window=30).mean()
df['EMA'] = df['Adj Close'].ewm(span=30, adjust=False).mean()

start_date = '2020-04-01'
end_date = '2020-09-24'
filtered_df = df.loc[start_date:end_date]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df.index, filtered_df['Adj Close'], label='Adj Close', color='blue', alpha=0.5)
plt.plot(filtered_df.index, filtered_df['30_SMA'], label='30 Day SMA', color='red')
plt.plot(filtered_df.index, filtered_df['EMA'], label='Exponential Moving Average', color='orange')

plt.title('Alphabet Inc. Adj Close, SMA, and EMA')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 16. Write a Pandas program to create a scatter plot of the trading volume/stock prices of Alphabet Inc. stock between two specific dates.

plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Close'], filtered_df['Volume'], alpha=0.5, color='purple')
plt.title('Alphabet Inc. Stock Price vs Trading Volume')
plt.xlabel('Stock Price (USD)')
plt.ylabel('Trading Volume')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 17. Write a Pandas program to create a plot to visualize daily percentage returns of Alphabet Inc. stock price between two specific dates.

df['Daily Return'] = df['Adj Close'].pct_change() * 100

start_date = '2020-04-01'
end_date = '2020-09-24'
filtered_df = df.loc[start_date:end_date]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df.index, filtered_df['Daily Return'], color='purple', lw=1.5)
plt.axhline(0, color='black', linestyle='--', alpha=0.5)
plt.title('Alphabet Inc. Daily Percentage Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

```python
# 18. Write a Pandas program to plot the volatility over a period of time of Alphabet Inc. stock price between two specific dates.

start_date = '2020-04-01'
end_date = '2020-09-24'
filtered_df = df.loc[start_date:end_date].copy()

filtered_df['Daily Return'] = filtered_df['Close'].pct_change()
filtered_df['Volatility'] = filtered_df['Daily Return'].rolling(window=7).std() * np.sqrt(7)

plt.figure(figsize=(12, 6))
plt.plot(filtered_df.index, filtered_df['Volatility'], color='orange', lw=2)
plt.title('Alphabet Inc. Stock Price Volatility (7-Day Rolling)')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 19. Write a Pandas program to create a histogram to visualize daily return distribution of Alphabet Inc. stock price between two specific dates.

df['Daily Return'] = df['Adj Close'].pct_change()

start_date = '2020-04-01'
end_date = '2020-09-24'
filtered_df = df.loc[start_date:end_date]

plt.figure(figsize=(10, 6))
filtered_df['Daily Return'].hist(bins=50, color='teal', edgecolor='black', grid=False)

plt.title('Distribution of Alphabet Inc. Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='dashed', linewidth=1)
plt.show()
```
