```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. Write a Python programming to display a bar chart of the popularity of programming Languages.
#Sample data:
# Programming languages: Java, Python, PHP, JavaScript, C#, C++
# Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(languages, popularity, color='blue')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#2. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
plt.barh(languages, popularity, color='green')
plt.ylabel("Programming Languages")
plt.xlabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#3. Write a Python programming to display a bar chart of the popularity of programming Languages. Use uniform color.
plt.bar(languages, popularity, color='royalblue')
plt.gca().invert_yaxis()
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages", y=-0.15)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#4. Write a Python programming to display a bar chart of the popularity of programming Languages. Use different color for each bar.
colors = ['red', 'black', 'green', 'blue', 'yellow', 'cyan']
plt.bar(languages, popularity, color=colors)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#5. Write a Python programming to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its
# popularity (float value).
bars = plt.bar(languages, popularity, color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.3, f'{yval:.6f}', ha='center', va='bottom')

plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#6. Write a Python programming to display a bar chart of the popularity of programming Languages. Make blue border to each bar.
bars = plt.bar(languages, popularity, color='skyblue', edgecolor='blue', linewidth=1.5)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#7. Write a Python programming to display a bar chart of the popularity of programming Languages. Specify the position of each bar plot.
positions = [1, 2, 4, 7, 8, 10]
plt.bar(positions, popularity, color='skyblue')
plt.xticks(positions, languages)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages (Specified Positions)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#8. Write a Python programming to display a bar chart of the popularity of programming Languages. Select the width of each bar and their positions.
positions = [1, 2, 5, 8, 10, 13]
widths = [0.1, 0.2, 0.5, 2.0, 0.2, 0.7]
plt.bar(positions, popularity, width=widths, color='blue', zorder=3)
plt.xticks(positions, languages)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages (Specified Widths & Positions)")
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='lightcoral', zorder=0)
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='grey', zorder=0)
plt.show()
```

```python
#9. Write a Python programming to display a bar chart of the popularity of programming Languages. Increase bottom margin.
plt.bar(languages, popularity, color='skyblue',zorder=3)

plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages")
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='lightcoral', zorder=0)
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='grey', zorder=0)
plt.subplots_adjust(bottom=0.4)
plt.show()
```

```python
#10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
groups = 5
means_men = (22, 30, 35, 35, 26)
means_women = (25, 32, 30, 35, 29)

fig, ax = plt.subplots()
index = np.arange(groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_men, bar_width,
alpha=opacity,
color='g',
label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
alpha=opacity,
color='r',
label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()

plt.tight_layout()
plt.show()
```
