```python
#17. Write a Pandas program to get the items which are not common of two given series.
sr1=pd.Series([1,2,3,4,5])
sr2=pd.Series([2,4,6,8,10])
print("Series 1:\n",ser1)
print("Series 2:\n",ser2)
print("\nItems of a given series not present in another given series:")
union = pd.Series(np.union1d(sr1, sr2))
intersect = pd.Series(np.intersect1d(sr1, sr2))
result = union[~union.isin(intersect)]
print(result)
```

```python
#28. Write a Pandas program to get the day of month, day of year, week number and day
#of week from a given series of date strings.
from dateutil.parser import parse

date_series = pd.Series([
    '01 Jan 2015', 
    '10-02-2016', 
    '20180307', 
    '2014/05/06', 
    '2016-04-12', 
    '2019-04-06T11:20'
])
print(date_series)

dates = date_series.map(lambda x: parse(x))

print("Day of month:")
print(dates.dt.day.tolist())

print("\nDay of year:")
print(dates.dt.dayofyear.tolist())

print("\nWeek number:")
print(dates.dt.isocalendar().week.tolist())

print("\nDay of week:")
print(dates.dt.day_name().tolist())
```

```python
#34. Write a Pandas program to compute the autocorrelations of a given numeric series. From Wikipedia: Autocorrelation,
# also known as serial correlation, is the correlation of a signal with a delayed copy of itself as a function of delay. 
# Informally, it is the similarity between observations as a function of the time lag between them.

series = pd.Series(np.arange(15)+ np.random.normal(10, 5, 15))

autocorrelations = [round(series.autocorr(lag=i), 2) for i in range(1, 11)]

print("Original series:")
print(series)

print("\nAutocorrelations of the said series:")
simple_list = [float(x) for x in autocorrelations]
print(simple_list)
```
