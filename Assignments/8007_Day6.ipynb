```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. Write a Python programming to display a bar chart of the popularity of programming Languages.
# Sample data:
# Programming languages: Java, Python, PHP, JavaScript, C#, C++
# Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(languages, popularity, color='blue', zorder=3)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='lightcoral', zorder=0)
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='grey', zorder=0)
plt.show()
```

```python
#14.Write a Python program to create a stacked bar plot with error bars.
#Note: Use bottom to stack the women?s bars on top of the men?s bars.

groups = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)
men_std = (4, 3, 4, 1, 5)
women_std = (3, 5, 2, 3, 3)
ind = np.arange(len(groups))
plt.figure()
plt.bar(ind, men_means, width=0.5, yerr=men_std, color='red', label='Men', capsize=5, zorder=3)
plt.bar(ind, women_means, width=0.5, bottom=men_means, yerr=women_std, color='green', label='Women', capsize=5, zorder=3)
plt.title('Scores by Group and Gender')
plt.xlabel('Group')
plt.ylabel('Scores')
plt.xticks(ind, groups)
plt.legend()
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='grey', alpha=0.3, zorder=0)
plt.show()
```

```python
#15. Write a Python program to create a horizontal bar chart with differently ordered colors.
# Note: Use bottom to stack the women?s bars on top of the men?s bars.

students = ['Student-1', 'Student-2', 'Student-3']

languages = [['Language','Science','Math'], ['Science','Math','Language'],
             ['Math','Language','Science']]

numbers = [{'Language':75, 'Science':88, 'Math':96}, 
           {'Language':71, 'Science':95, 'Math':92}, 
           {'Language':75, 'Science':90, 'Math':89}]

colors = {'Language':'red', 'Science':'blue', 'Math':'green'}

plt.figure()
added_labels = set()

for i, student in enumerate(students):
    left_offset = 0
    
    for subject in languages[i]:
        label = subject if subject not in added_labels else ""
        plt.barh(student, numbers[i][subject], left=left_offset, 
                 color=colors[subject], edgecolor='black', zorder=3, label=label)
        added_labels.add(subject)
        left_offset += numbers[i][subject]
        
plt.title('Subject Scores by Student')
plt.xlabel('Total Score')
plt.grid(axis='x', linestyle='--', alpha=0.7, zorder=0)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
#16. Write a Python program to create stack bar plot and add label to each section.

people = ('G1','G2','G3','G4','G5','G6','G7','G8')

data = np.array([[ 3.4, 7.7, 6.4, 10.5, 7.5, 7.1, 12.7, 3.4],
[ 11.2, 5.0, 6.6, 12.3, 7.4, 6.7, 7.2, 3.6],
[ 3.9, 4.7, 11.7, 4.6, 12.9, 4.6, 11.1, 8.5],
[ 4.2, 12.7, 11.3, 9.0, 10.4, 10.9, 3.9, 8.8]])

colors_list = ['red', 'green', 'white', 'purple']

plt.figure()
left_offset = np.zeros(len(people))

for i in range(len(data)):
    bars = plt.barh(people, data[i], left=left_offset, color=colors_list[i], 
                    edgecolor='black', zorder=3)
    for j, bar in enumerate(bars):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_y() + bar.get_height()/2, 
                 f'{int(data[i][j])}%', ha='center', va='center')
    left_offset += data[i]
    
plt.title('Group Performance Segments')
plt.xlabel('Scores')
plt.show()
```
