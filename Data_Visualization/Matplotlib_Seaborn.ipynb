# Data Visualization Using Matplotlib and Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

## 1. Line Plot - Matplotlib and Seaborn

```python
# 1. Line Plot - Matplotlib

plt.figure(figsize=(6,4))
plt.plot(df.index,df["tip"],marker="o",color="black")
plt.title("Matplotlib Line Plot - Tip across records")
plt.xlabel("Record Index")
plt.ylabel("Tips")
plt.show()
```

```python
# 2. Line Plot - Seaborn

plt.figure(figsize=(6,4))
sns.lineplot(x=df.index, y=df["tip"],marker="o",color="black")
plt.title("Seaborn Line Plot - Tip across records")
plt.xlabel("Record Index")
plt.ylabel("Tips")
plt.show()
```

## 2. Bar Plot - Matplotlib and Seaborn

```python
# 1. Bar Plot - Matplotlib

avg_bill_day = df.groupby("day") ["total_bill"].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_bill_day.index,avg_bill_day.values,color="black")
plt.title("Matplotlib Bar Plot - Avg Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()
```

```python
# 2. Bar Plot - Seaborn

plt.figure(figsize=(6,4))
sns.barplot(x="day",y="total_bill",data=df,color="black",errorbar=None)
plt.title(" Seaborn Bar Plot - Avg Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()
```

## 3. Histogram - Matplotlib and Seaborn

```python
# 1. Histogram - Matplotlib

plt.figure(figsize=(6,4))
plt.hist(df["total_bill"],bins=7,color="black",edgecolor="white")
plt.title("Matplotlib Histogram - Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Histogram - Seaborn

plt.figure(figsize=(6,4))
sns.histplot(df["total_bill"],bins=7,color="black",edgecolor="white")
plt.title("Seaborn Histogram - Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

## 4. Box Plot - Matplotlib and Seaborn

```python
# 4. Box Plot - Matplotlib

tip_by_day = [df[df["day"]==d]["tip"]for d in df["day"].unique()]
plt.figure(figsize=(6,4))
plt.boxplot(tip_by_day, tick_labels = df["day"].unique())
plt.title("Matplotlib Box Plot - Tip by Day")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
# 4. Box Plot - Seaborn

plt.figure(figsize=(6,4))
sns.boxplot(x="day",y="tip",data=df,color="pink")
plt.title("Seaborn Box Plot - Tip by Day")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

## 5. Scatter Plot - Matplotlib and Seaborn

```python
# 5. Scatter Plot - Matplotlib

plt.figure(figsize=(5,5))
plt.scatter(df["total_bill"],df["tip"],color="black")
plt.title("Matplotlib Scatter Plot - Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# 5. Scatter Plot - Seaborn

plt.figure(figsize=(5,5))
sns.scatterplot(x="total_bill",y="tip",data=df,color="black")
plt.title("Seaborn Scatter Plot - Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

## 6. Heatmap - Seaborn

```python
# 1. Heatmap - Seaborn

corr = df[["total_bill","tip","size"]].corr()
plt.figure(figsize=(5,5))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Seaborn Heatmap - Correlation Matrix")
plt.show()
```
