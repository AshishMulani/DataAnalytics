```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
avg_tip_by_day = df.groupby("day") ["tip"].mean()
corr = df[["total_bill","tip","size"]].corr()
```

## 1. Traditional/Functional pyplot Style

```python
# Line
plt.subplot(2,3,1)
plt.plot(df.index,df["total_bill"],color="#F3F589")
plt.title("Line Plot")

# Bar 
plt.subplot(2,3,2)
plt.bar(avg_tip_by_day.index, avg_tip_by_day.values,color="#B489F5")
plt.title("Bar Plot")

# Histogram
plt.subplot(2,3,3)
plt.hist(df["total_bill"],bins=10,color="#F589E3")
plt.title("Histogram")

# Box Plot
plt.subplot(2,3,4)
plt.boxplot(df["tip"])
plt.title("Box Plot")

# Scatter Plot
plt.subplot(2,3,5)
plt.scatter(df["total_bill"],df["tip"],color="#89BBF5")
plt.title("Scatter Plot")


# Heatmap
plt.subplot(2,3,6)
plt.imshow(corr)
plt.title("Heatmap")
plt.colorbar()

plt.tight_layout()
plt.show()
```

## 2. Object-Oriented Style

```python
fig, axes = plt.subplots(2,3,figsize=(12,8))

# Lines 
axes[0,0].plot(df.index,df["total_bill"])
axes[0,0].set_title("Line Plot")

#Bar
axes[0,1].bar(avg_tip_by_day.index,avg_tip_by_day.values)
axes[0,1].set_title("Bar Plot")

#Histogram
axes[0,2].hist(df["total_bill"],bins=10)
axes[0,2].set_title("Histogram")

# Box Plot
axes[1,0].boxplot(df["tip"])
axes[1,0].set_title("Box Plot")

# Scatter Plot
axes[1,1].scatter(df["total_bill"],df["total_bill"])
axes[1,1].set_title("Scatter Plot")

#Heatmap
img=axes[1,2].imshow(corr)
axes [1,2].set_title("Heatmap")

plt.tight_layout()
plt.show()
```
