```python
import numpy as np
```

```python
# arange - excludes last index

# np.arange(start,stop,step)
arr = np.arange(0,10,2) # 0 - start index, 10 - stop index, 2 - gap of 2 in given range
arr
```

```python
arr = np.arange(0,100,2) 
arr
```

```python
arr = np.arange(0,100,50) 
arr
```

```python
# linspace - include last index

# np.linspace(start,stop,num)
arr = np.linspace(0,10,5) # 0 - start at no. zero, 10 - stop at no. 10, 5 - divide equally the given range in equal parts(num of elements)
arr
```

```python
arr = np.linspace(0,100,5) 
arr
```

```python
arr = np.linspace(0,100,2) 
arr
```

```python
arr = np.linspace(0,100,3) 
arr
```

#### Basic Usage

```python
import numpy as np
a = np.arange(0,10,2)
b = np.linspace(0,10,5)
print("arange (0,10,2):",a)
print("linspace (0,10,5):",b)
```

#### Float Behavior

```python
print("\narange with float step (miss endpoint):")
print(np.arange(0,1,0.2))
print("\nlinspace with exact number of points :")
print(np.linspace(0,10,6))
```

#### Endpoint Control

```python
print("\nlinspace without endpoint:") 
print(np.linspace(0,10,5,endpoint=False)) # To exclude end value
```

#### Length Comparison

```python
print("\nNumber of elements :")
print("len(arange) :",len(np.arange(0,10,2)))
print("len(linspace) :",len(np.linspace(0,10,5)))
```

#### Precision Demonstration

```python
print("\nPrecision check :")
arr = np.arange(0,1,0.1)
lin = np.linspace(0,1,10)
print("arange:", arr)
print("linspace:",lin)
```
