```python
import numpy as np
```

```python
#step1:create array
arr=np.arange(55,65)
print("Original Array :\n",arr)
```

```python
#step2:Apply condition(vectorized)
mask=arr>58
print("\nCondition (arr>58):",mask)
```

```python
#step3:Apply Boolean mask
boolean_arrr=arr>58
boolean_arr
```

```python
#step4:Filter values
filtered=arr[mask]
print("\filtered values:",filtered)
```

```python
#direct filtering
print("Direct filtering (best practice):",arr[arr>58])
```

```python
#multiple conditions
print("\nValues between 57 and 62:",arr[(arr>57)&(arr<63)])
```

#### Shallow Copy and Deep Copy

```python
a = np.array([10,20,30,40,50,60])
```

```python
# Shallow Copy(View)
b = a[0:5]
print("Original a :",a)
print("Shallow Copy b :",b)
```

```python
b[0]=999
print("\nAfter modifying b :")
print("a:",a)
print("b:",b)
```

```python
# Deep Copy
c=a.copy()
```

```python
c[1]=888
print("\nAfter modifying c:")
print("a:",a)
print("c:",c)
```
