```python
import numpy as np
```

```python
# Create
arr = np.array([1,2,3,4])
lst=[1,2,3,4]
arr
```

```python
# Add 
print(arr+5)
print([x+5 for x in lst])
```

```python
# (Time) - Add 5 and measure time for numpy
import time
start_time = time.time()
print(arr+5)
end_time = time.time()
execution_time = end_time - start_time
print(f"\nExecution Time : {execution_time} seconds")
print([x+5 for x in lst])
```

```python
# (Time) - Add 5 and measure time for list
start_time = time.time()
print([x+5 for x in lst])
end_time = time.time()
execution_time = end_time - start_time
print(f"\nExecution Time : {execution_time} seconds")
```

```python
# Multiply
print(arr*5)
print([x*5 for x in lst])
```

```python
# Square
print(arr**2)
print([x**2 for x in lst])
```

```python
# Mean
print(np.mean(arr))
print(sum(lst)/len(lst))
```

```python
# Even Filter
print(arr[arr%2 == 0])
print([x for x in lst if x % 2 == 0])
```

```python
# 2D Data
print(np.array([[1,2],[3,4]]))
print([[1,2],[3,4]])
```

```python
# Add Two - NumPy
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
```

```python
# Add Two - List 
a = [1,2,3]
b = [4,5,6]
print([x+y for x, y in zip(a,b)])

# zip(a,b) -- read the same element of array and array b at the same time, when processing a[0], also process b[0]
# for x,y in zip(a,b) -- a[0] - x and b[0] - y
# x+y for x,y in zip(a,b) - Add a and y, which means a[0] and b[0], next time a[1] and b[1], becoz we have a "for" loop
```

```python
# Conditional Replace
arr[arr>2] = 0
print(arr)

lst=[1,2,3,4]
print([0 if x>2 else x for x in lst])
```
