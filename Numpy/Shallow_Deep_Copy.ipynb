#### Shallow Copy and Deep Copy

```python
import numpy as np
a = np.array([10,20,30,40,50,60])
```

```python
# Shallow Copy(View)
b = a[0:5]
print("Original a :",a)
print("Shallow Copy b :",b)
```

```python
b[0]=999
print("\nAfter modifying b :")
print("a:",a)
print("b:",b)
```

```python
# Deep Copy
c=a.copy()
```

```python
c[1]=888
print("\nAfter modifying c:")
print("a:",a)
print("c:",c)
```
