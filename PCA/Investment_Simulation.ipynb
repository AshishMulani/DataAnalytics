```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters

initial_investment =10000    # initial investment in dollars
mean_return = 0.15           # Expected annual return (15%)
std_dev = 0.15               # std dev of returns (15%)

num_simulations = 1000       # no. of monte carlo simulations
time_horizon = 10            # in years

# Monte Carlo Simulation
simulated_end_values = []

for _ in range(num_simulations):
    # initialize the value of the investment
    investment_value = initial_investment

    for _ in range(time_horizon):
        # Generate a random annual return
        random_return = np.random.normal(mean_return, std_dev)

        # update the investment value for the year
        investment_value *= (1+random_return)

    simulated_end_values.append(investment_value)

# Results
mean_end_value = np.mean(simulated_end_values)
risk_of_loss = np.mean(np.array(simulated_end_values) < initial_investment)

print(f"Mean end Value: ${mean_end_value:.2f}")
print(f"Probability of Loss: {risk_of_loss * 100:.2f}%")
```

```python
# Plotting the Distribution
plt.hist(simulated_end_values, bins=30, edgecolor='k')
plt.title('Simulated Portfolio End Values After 5 Years')
plt.xlabel('Portfolio Value ($)')
plt.ylabel('Frequency')
plt.show()
```
