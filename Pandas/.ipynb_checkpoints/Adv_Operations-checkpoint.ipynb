## loc and iloc

```python
import numpy as np
import pandas as pd
```

```python
# Data

data = {
    "Player":["Virat","Rohit","MS Dhoni"],
    "Match 1": [44,55,30],
    "Match 2": [67,72,40],
    "Match 3": [82,60,50]
}

df = pd.DataFrame(data).set_index("Player")
# print("DataFrame:\n",df)
df
```

```python
# First Row

print("\nFirst row:")
print("loc:",df.loc["Virat"].values)
print("iloc:",df.iloc[0].values)
```

```python
# Second Row

print("\nSecond row:")
print("loc:",df.loc["Rohit"].values)
print("iloc:",df.iloc[1].values)
```

```python
# Specific Cell

print("\nSpecific Cell (Virat,Match 2):")
print("loc:",df.loc["Virat","Match 2"])
print("iloc:",df.iloc[0,1])
```

```python
# Multiple rows
print("\nMultiple rows:")
print("loc:\n",df.loc[["Virat","Rohit"]])
print("iloc:\n",df.iloc[0:2])
```

```python
# Subset

print("\nSubset:")
print("loc:\n",df.loc[["Virat","MS Dhoni"],["Match 1", "Match 3"]])
print("iloc:\n",df.iloc[[0,2],[0,2]])
```

```python
# Slicing 

print("\nSlicing :")
print("loc:\n",df.loc["Virat":"Rohit"])
print("iloc:\n",df.iloc[0:2])
```

## Advanced Operations

```python
# Sample Data

df = pd.DataFrame({

    "Employee" : ["Amit","Neha","Raj","Neha","Amit"],
    "Department": ["IT","HR","IT","HR","Finance"],
    "Salary": [50000,60000,np.nan,65000,70000]
})
df
```

```python
# Indexing

df_indexed=df.set_index("Employee")
df_indexed
```

```python
# Filtering

df[df["Salary"]> 60000]
```

```python
# Check Missing

df.isnull()
```

```python
# Fill Missing
df.fillna(3)
```

```python
# Drop Missing
df.dropna()
```

```python
# Grouping
print("\nAverage Salary by Department:\n")
df.groupby("Department")["Salary"].mean()
```

```python
# Merge (Join)

df_bonus = pd.DataFrame({
    "Employee": ["Amit","Neha","Raj"],
    "Bonus":[5000,7000,4000]
})
merged = pd.merge(df,df_bonus,on="Employee")
print("\nMerged DataFrame:\n",merged)
merged
```

```python
# Merge (Join)
#drops the row who's value is not given or adds only common rows
df_bonus = pd.DataFrame({
    "Employee": ["Amit","Neha"],
    "Bonus":[5000,7000]
})
merged = pd.merge(df,df_bonus,on="Employee")
print("\nMerged DataFrame:\n",merged)
merged
```

```python
# Concat 
df_new = pd.DataFrame({
    "Employee" :["Kiran"],
    "Department":["IT"],
    "Salary":[55000]
})
concat_df = pd.concat([df,df_new],ignore_index=True)
print("\nConcatenated DataFrame :\n", concat_df)
concat_df
```

```python
# Concat 
# assigns null value for the row where value is not given
df_new = pd.DataFrame({
    "Employee" :["Kiran"],
    "Department":["IT"]
   
})
concat_df = pd.concat([df,df_new],ignore_index=True)
print("\nConcatenated DataFrame :\n", concat_df)
concat_df
```
