```python
import pandas as pd
import seaborn as sns
```

```python
#Load tips dataset

df=sns.load_dataset("tips")
df.head()
```

```python
# 1.Boolean Filtering

df[df['total_bill']>20].head()
```

```python
# 2.Multiple Conditions

df[(df['total_bill'] > 20) & (df['tip']<3)].head()
```

```python
# 3. .loc[] Filtering + Select Columns

df.loc[df['size'] ==2,['total_bill','tip','sex','size']].head()
```

```python
# 4.query

df.query("total_bill > 20 and tip < 3").head()
```

```python
# 5. isin()

df[df['day'].isin(['Fri'])].head()
```

```python
# 6. between()

df[df['total_bill'].between(15,25)].head()
```

```python
# 7. String Filter

df[df['sex'].astype(str).str.contains("Male")].head()
```

```python
# 8. Top Values

df.nlargest(10,'tip')
```
