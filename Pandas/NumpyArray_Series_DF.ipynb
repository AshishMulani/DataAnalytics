```python
import numpy as np
import pandas as pd
```

## NumPy Arrays

```python
# Data

companies = np.array(["TCS", "Infosys","Wipro"])
sales = np.array([100,200,300])
profit = np.array([20,50,70])
```

```python
print("========== NUMPY ARRAYS ==========")
print("Companies : ",companies)
print("Sales :",sales)
print("Profit :",profit)
```

```python
# Accessing Using Index

print("\nUsing NumPy :")
print("Infosys Sales :",sales[1])
print("Infosys Profit :",profit[1])
```

## Pandas Series

```python
sales_s = pd.Series(sales,index=companies)
profit_s = pd.Series(profit,index=companies)
print("Sales Series :\n",sales_s)
print("\nProfit Series :\n",profit_s)
```

```python
# Accessing Using Label

print("\nUsing Series :")
print("Infosys Sales :",sales_s["Infosys"])
print("Infosys Profit :",profit_s["Infosys"])
```

## Exercise

```python
# We want to store student prn, name, and mobile no. for 3 students. 
# Use NumPy arrays and Pandas Series. Print 3rd student's name and 2nd student's mobile no.
```

```python
# NumPy Array

prn = np.array([100,200,300])
name = np.array(["Ashish", "Abhinay","Akshay"])
mobile = np.array([873824745,504558766,7065786654])
print("========== Student Details ==========")
print("prn : ",prn)
print("name :",name)
print("mobile :",mobile)
```

```python
print("3rd student's name :", name[2])
print("2nd students's mobile no. :", mobile[1])
```

```python
# Pandas Series

name_s = pd.Series(name,index=prn)
mobile_s = pd.Series(mobile,index=prn)
print("Sales Series :\n",name_s)
print("\nProfit Series :\n",mobile_s)
```

```python
print("\nUsing labels :")
print("3rd student's name :",name_s[300])
print("2nd stident's mobile no :",mobile_s[200])
print()
print("Using iloc :")
print("3rd student's name :",name_s.iloc[2])
print("2nd stident's mobile no :",mobile_s.iloc[1])
```

## DataFrame

```python
df = pd.DataFrame({
     "Sales":sales,
     "Profit": profit },
      index = companies)
print(df)
```

```python
# Access row and multiple columns

print("\nUsing DataFrame:")
print(df.loc["Infosys",["Sales","Profit"]])
```

```python
# Add new column

df["Margin"] = df["Profit"] / df["Sales"]*100
print("\nWith Margin:\n",df)
```

```python
# Filter multiple columns

print("\nHigh Profit Companies:\n", df[df["Profit"]>40])
```

```python
# Combined condition

print("\nHigh Sales and Profit:\n",
     df[(df["Sales"]>150) & (df["Profit"] > 40)])
```
