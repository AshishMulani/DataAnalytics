## 1-sample test

```python

import pandas as pd
from scipy import stats
```

```python
df=pd.read_csv('tips.csv')
df.head()
```

```python
our_column=df['tip']
our_column=our_column.sample(n=30, replace=False)

meu=3   #we assume that avg tip is $3 and want to test it

# H0: avg tip = $3
# H1: avg tip <> $3

t_stat, p_value= stats.ttest_1samp(our_column, meu)
print('p_value: ',p_value)
```

```python
alpha=0.05
if p_value < alpha:
    print('reject H0')

else:
    print("Do not reject H0")
```

## 2-Sample/independent t-test

```python
male_tips= df[df['sex']=='Male']['tip']
female_tips= df[df['sex']=='Female']['tip']

male_tips= male_tips.sample(n=30, replace=False)
female_tips= female_tips.sample(n=30, replace=False)

t_stat, p_value= stats.ttest_ind(male_tips, female_tips)
print('p_value: ',p_value)
```

```python
if p_value < alpha:
    print('reject H0')

else:
    print("Do not reject H0")
```

## Paired t-test

```python
df=pd.read_csv('hypothermia.csv')
df= df[['t.1', 't.2']].dropna()
df.head()
```

```python
sample_df= df.sample(n=30, replace=False)
t_stat, p_value= stats.ttest_rel(sample_df['t.1'], sample_df['t.2'] )
print('p_value: ',p_value)
```

```python
if p_value < alpha:
    print('reject H0')
else:
    print("Do not reject H0")

# H0: Treatment is not effective
# H1: Treatment is effective
```
