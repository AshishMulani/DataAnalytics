```python
from pulp import *
```

```python
# Create variables for Chairs and Tables
model = LpProblem("FactoryProductionProblem",LpMaximize)
```

```python
# Create variables for chairs and tables
# Pramateres : Name, Lower Limit, Upper Limit, Data type

chairs = LpVariable("Chairs",0,None,LpInteger) # No. of chairs to produce
tables=LpVariable("Tables",0,None,LpInteger) # No. of tables to produce
```

```python
# Create maximize objective function(profit)

model += 300*chairs+1000*tables,"Profit"
```

```python
# Create constraints

model +=5*chairs + 10*tables<=400,"Wood Contraints" # Wood availability constraint
model +=2*chairs + 6*tables<=300,"Time Contraints" # Worker time constraint
```

```python
model.solve()
```

```python
# Each of the variables is printed eith its resolved optium value 

for v in model.variables():
    print(v.name,"=", v.varValue)
print("Total Profit:",value(model.objective))
```

# Minimization Objective (Bank Loan Example)

```python
emi_bank_a = 2076
emi_bank_b = 3227
emi_bank_c = 4707
total_loan_required=10

model=LpProblem("LoanDistributionProblem",LpMinimize)
x1=LpVariable("Loan_from_Bank_A",0,None,LpContinuous)
x2=LpVariable("Loan_from_Bank_B",0,None,LpContinuous)
x3=LpVariable("Loan_from_Bank_C",0,None,LpContinuous)

model +=emi_bank_a * x1+emi_bank_b * x2+emi_bank_c * x3,"Total_Monthly_EMI"

model += x1 + x2 + x3 == total_loan_required,"Total_Loan_Constraint"
model += x1 >=2,"Min_Loan_Bank_A"
model += x3 <=4,"Max_Loan_Bank_C"
model.solve()

for v in model.variables():
    print(v.name,"=",v.varValue)
print(f'Minimum Total Monthly EMI: Rs {value(model.objective)}')
```
