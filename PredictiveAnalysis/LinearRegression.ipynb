```python
import pandas as pd
```

```python
df=pd.read_csv("Salary.csv")
x=df[["YearsExperience"]]
y=df["Salary"]
```

```python
#step2: split data into training and testing
#!pip install scikit-learn
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
#step3: Fit simple linear regression to Training data
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
#step4: make prediction
y_pred= model.predict(x_test)
```

```python
#step5: Visualize test set results
import matplotlib.pyplot as plt
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train, model.predict(x_train), color='blue') #same regression line
plt.title('Salary Vs Experience (Test Set)')
plt.xlabel("Years of Experience")
plt.ylabel('Salary')
plt.show()
```

```python
#step6: Make new prediction
new_x_values_df = pd.DataFrame({'YearsExperience': [1,3,5,8,12,15,20,25]})
new_y_pred = model.predict(new_x_values_df)
print('\nPredictions for new Experience Values (1,3,5,8,12,15,20,25):\n',new_y_pred)
```

```python
#step7: Model Evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae= mean_absolute_error(y_test, y_pred)
mse= mean_squared_error(y_test, y_pred)
r2= r2_score(y_test, y_pred)

print('\nModel Evaluation Metrics:')
print('\nMean Absolute Error (MAE):',mae)
print('\nMean Squared Error (MSE):',mse)
print('\nR2 Score:',r2)
```
