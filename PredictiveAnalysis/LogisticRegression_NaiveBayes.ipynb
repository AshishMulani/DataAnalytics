## Logistic Regression 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Given model parameters
b0=-10
b1=0.2
```

```python
# Dog Weights(Our Sample Data)
weights=np.array([20,30,40,50,60,70])
```

```python
# Logit Values
z=b0+b1*weights
```

```python
# Sigmoid Function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
# Probabilities
probs = sigmoid(z)
probs
```

```python
# Print Table

print("Weigt Logit(z) Probability Result")
for w, logit,p in zip(weights,z,probs):
    result="Obese" if p>=0.5 else "Not Obese"
    print(f"{w:6} {logit:8.2f} {p:10.4f} {result}")
    
```

```python
# Create Smooth Curve
x_vals=np.linspace(0,80,200)
z_vals=b0+b1*x_vals
y_vals=sigmoid(z_vals)
```

```python
# Plot
plt.figure()
plt.plot(x_vals,y_vals,color='red') #sigmaoid curve
plt.scatter(weights,probs,color='black')#actual points

# Decision Boundary Line
plt.axhline(y=0.5,color='darkblue')
plt.xlabel("Dog Weight(kg)")
plt.ylabel("Probability of Obesity")
plt.title("Logistic Regression (Dog Obesity)")
plt.show()
```
