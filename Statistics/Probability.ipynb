```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df = df[['gender','survived']]
df.dropna(inplace=True)
total=len(df)
print(f"Total Passengers:{total}")
```

```python
# Marginal Probability of Gender

total_female = len(df[df['gender']=='female'])
print(f"Total female : {total_female}")
total_male=len(df[df['gender'] == 'male'])
print(f"Total male : {total_male}")
p_female = total_female/total
p_male = total_male/total
```

```python
# Marginal Probability of Survival

total_survived = len(df[df['survived']==1])
print(f"Total survived : {total_survived}")
total_did_not_survived=len(df[df['survived'] == 0])
print(f"Total did not survived : {total_did_not_survived}")
```

```python
p_survived = total_survived/total
p_not_survived = total_did_not_survived/total

print("\nMarginal Probabilities -")
print(f"P(gender =female):{p_female:.3f}")
print(f"P(gender = male):{p_male:.3f}")
print(f"survived = 1):{p_survived:.3f}")
print(f"survived = 0):{p_not_survived:.3f}")
```

```python
# Joint Probabilities

p_female_survived = len(df[(df['gender'] == 'female') & (df['survived'] == 1)])/total
p_female_not_survived = len(df[(df['gender'] == 'female') & (df['survived'] == 0)])/total
p_male_survived = len(df[(df['gender'] == 'male') & (df['survived'] == 1)])/total
p_male_not_survived = len(df[(df['gender'] == 'male') & (df['survived'] == 0)])/total   

print("\nJoint Probabilities -")
print(f"P(female, survived=1):{p_female_survived:.3f}")
print(f"P(female, survived=0):{p_female_not_survived:.3f}")
print(f"P(male, survived=1):{p_male_survived:.3f}")
print(f"P(male, survived=0):{p_male_not_survived:.3f}")
```

```python
# Conditional Probabilities
# P (Survive | Female)=p(Survive and Female) / P(Female)

p_survived_given_female = p_female_survived/p_female
print(f"P(survived=1 | gender = female): {p_survived_given_female :.3f}")

p_survived_given_male = p_male_survived/p_male
print(f"P(survived=1 | gender = male): {p_survived_given_male :.3f}")

```

### 1. What is the probability that a randomly selected person in the dataset has diabetes? (Marginal Probability)


```python
df=pd.read_csv("diabetes.csv")
df.dropna(inplace=True)
df
```

```python
total=len(df)
print(f"Total Persons:{total}")
```

```python
#Marginal Probability

total_dibetic = len(df[df['Outcome']==1])
print(f"Total Dibetic : {total_dibetic}")

total_did_not_dibetic=len(df[df['Outcome'] == 0])
print(f"Total did not Dibetic : {total_did_not_dibetic}")
```

```python
p_dibetic = total_dibetic/total
p_not_dibetic = total_did_not_dibetic/total

print(f"(Dibetic = 1):{p_dibetic:.3f}")
print(f"(Dibetic = 0):{p_not_dibetic:.3f}")
```

### 2. What is the probability that a person has diabetes and is obese (BMI>30)? (Joint Probability of Independent Events)

```python
# Joint Probability of Independent Events

p_BMI_dibetic = len(df[(df['BMI']>30) & (df['Outcome'] == 1)])/total
p_BMI_not_dibetic = len(df[(df['BMI']>30) & (df['Outcome'] == 0)])/total   

print("\nJoint Probabilities -")
print(f"P(BMI>30, Dibetic=1):{p_BMI_dibetic:.3f}")
print(f"P(BMI>30, Dibetic=0):{p_BMI_not_dibetic:.3f}")
```

### 3. What is the probability that a person has diabetes given that their glucose level is high (eg. Glucose > 140)? (Joint Probability of Dependent Events / Conditional Probability)

```python
# Marginal Probability

total_highGlucose = len(df[df['Glucose']>140])
print(f"Total High Glucose Patients : {total_highGlucose}")

p_highGlucose=total_highGlucose/total
print(f"(Glucose>140):{p_highGlucose:.3f}")

# Joint Probability of Dependent Events 

p_glucose_dibetic = len(df[(df['Glucose']>140) & (df['Outcome'] == 1)])/total
p_glucose_not_dibetic = len(df[(df['Glucose']>140) & (df['Outcome'] == 0)])/total   

print("\nJoint Probabilities of Dependent Events-")
print(f"P(Glucose>140, Dibetic=1):{p_glucose_dibetic:.3f}")
print(f"P(Glucose>140, Dibetic=0):{p_glucose_not_dibetic:.3f}")

# Conditional Probability

# P (Dibetic | GLucose>140)=p(Dibetic and GLucose) / P(GLucose)
print("\nConditional Probability-")
p_dibetic_given_highGlucose = p_glucose_dibetic/p_highGlucose
print(f"P(Dibetic=1 | Glucose>140): {p_dibetic_given_highGlucose :.3f}")

p_not_dibetic_given_highGlucose = p_glucose_not_dibetic/p_highGlucose
print(f"P(Dibetic=0 | Glucose>140): {p_not_dibetic_given_highGlucose :.3f}")
```
