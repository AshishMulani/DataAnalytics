```python
import pandas as pd
import numpy as np
```

```python
df_2015 = pd.read_csv("marathon_results_2015.csv")
df_2016 = pd.read_csv("marathon_results_2016.csv")
df_2017 = pd.read_csv("marathon_results_2017.csv")
```

```python
# Concatenate the datasets into a single DataFrame
df = pd.concat([df_2015,df_2016,df_2017])

# Get the total mnumberv of rows in the original dataframe
original_row_count=len(df)

# Perform simple random sampling (selecting 1/10th of the records)
sample_df = df.sample(frac=0.1,random_state=42)

# Get the total number of rows in the sampled dataframe
sample_row_count = len(sample_df)

print(df.head())
print(sample_df.head())

# Print the number of rows in original and sampled dataframes

print(f"Number of rows in the original dataframes: {original_row_count}")
print(f"Number of rows in the sampled dataframes: {sample_row_count}")
```

```python
# Perform systematc sampling(picking every 7th record)
# ::7 - from the start to the end, pick every 7th record
# ,: - pick up all the columns
sample_df = df.iloc[::7, :]
sample_row_count = len(sample_df)

print("First few rows of the original dataframes:")
print(df.head())
print("\nFirst few rows of the sampled dataframes:")
print(sample_df.head())
```

```python
print(f"Number of rows in the original dataframes:{original_row_count}")
print(f"Number of rows in the sampled dataframes:{sample_row_count}")
```

### How good is our sampling?

```python
#original data - get the times: official time and pace
#official time- hh:mm:ss

df[['Hours','Minutes','Seconds']]= df['Official Time'].str.split(':', expand=True)

df['Hours']=df['Hours'].astype(int)
df['Minutes']=df['Minutes'].astype(int)
df['Seconds']=df['Seconds'].astype(int)

df['finish_time_in_seconds']=(df['Hours']*3600 + 
                             df['Minutes']*60 +
                             df['Seconds'])

#Pace-> hh:mm:ss
df[['Pace_Hours','Pace_Minutes','Pace_Seconds']]= df['Pace'].str.split(':', expand=True)

df['Pace_Hours']= df['Pace_Hours'].astype(int)
df['Pace_Minutes']= df['Pace_Minutes'].astype(int)
df['Pace_Seconds']= df['Pace_Seconds'].astype(int)

df['Pace_in_Seconds']=(df['Pace_Hours']*3600 + 
                             df['Pace_Minutes']*60 +
                             df['Pace_Seconds'])

#correlation
original_corr= df['finish_time_in_seconds'].corr(df['Pace_in_Seconds'])
print(f'Original correlation: ',original_corr)
```

```python
#Sample data - get the times: official time and pace
#official time- hh:mm:ss

sample_df[['Hours','Minutes','Seconds']]= sample_df['Official Time'].str.split(':', expand=True)

sample_df['Hours']=sample_df['Hours'].astype(int)
sample_df['Minutes']=sample_df['Minutes'].astype(int)
sample_df['Seconds']=sample_df['Seconds'].astype(int)

sample_df['finish_time_in_seconds']=(sample_df['Hours']*3600 + 
                             sample_df['Minutes']*60 +
                             sample_df['Seconds'])

#Pace-> hh:mm:ss
sample_df[['Pace_Hours','Pace_Minutes','Pace_Seconds']]= sample_df['Pace'].str.split(':', expand=True)

sample_df['Pace_Hours']= sample_df['Pace_Hours'].astype(int)
sample_df['Pace_Minutes']= sample_df['Pace_Minutes'].astype(int)
sample_df['Pace_Seconds']= sample_df['Pace_Seconds'].astype(int)

sample_df['Pace_in_Seconds']=(sample_df['Pace_Hours']*3600 + 
                             sample_df['Pace_Minutes']*60 +
                             sample_df['Pace_Seconds'])

#correlation
sample_corr= sample_df['finish_time_in_seconds'].corr(sample_df['Pace_in_Seconds'])
print(f'Sample correlation: ',sample_corr)
```

### Stratified Random Sampling

```python
#Find the total number of male and female athletes

total_males= len(df[df['M/F']=='M'])
total_females= len(df[df['M/F']=='F'])

#Calculate the proportionate number of male and female athletes for sample
sample_size= int(0.1*original_row_count)
sample_males= int(0.1*total_males)
sample_females= int(0.1*total_females)
print(sample_size, sample_males, sample_females)

#Generate proportionate samples of male and female athletes
male_sample= df[df['M/F']=='M'].sample(n=sample_males, random_state=42)
female_sample= df[df['M/F']=='F'].sample(n=sample_females, random_state=42)
print(len(male_sample), len(female_sample))

#Combine the male and female samples into a single dataframe
sample_df= pd.concat([male_sample, female_sample])

#Get the total number of rows in the sampled dataframe
sample_row_count= len(sample_df)
print("First few rows of the Original Dataframe: ")
print(df.head())

print("\nFirst few rows of the Sampled Dataframe: ")
print(sample_df.head())

print(f"\nNumber of rows in the Original Dataframe: {original_row_count}")
print(f"\nNumber of rows in the Sampled Dataframe: {sample_row_count}")
```

```python
#Find the Unique Countries
unique_countries= df['Country'].unique()
num_countries= len(unique_countries)
print(f'Unique Countries:\n {unique_countries}')
print(f'\n Number of Countries:\n{num_countries}')

#Randomly select half of the Countries
np.random.seed(42)
#choice() function in numpy generates a random 
selected_countries=np.random.choice(unique_countries, size=num_countries//2, replace=False) #replace=False means do not allow the same 
#country twice or more
#Example: np.random.choice(unique_countries, size=num_countries//2, replace=True) may give ['USA', 'USA']
print(f'Selected Countries:\n {selected_countries}')

#Create a new Dataframe containing all records of the selected countries
sample_df= df[df['Country'].isin(selected_countries)]

sample_row_count= len(sample_df)
print(f"\nNumber of rows in the Original Dataframe: {original_row_count}")
print(f"\nNumber of rows in the Sampled Dataframe: {sample_row_count}")
```
