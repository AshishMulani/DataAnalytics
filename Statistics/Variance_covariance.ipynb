```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('weight-height.csv')
df
```

```python
df_male = df[df['Gender'] =='Male']
df_female = df[df['Gender'] =='Female']
```

```python
corr_overall = df[['Height','Weight']].corr()
corr_male = df_male[['Height','Weight']].corr()
corr_female = df_female[['Height','Weight']].corr()
```

```python
cov_overall = df[['Height','Weight']].cov()
cov_male = df_male[['Height','Weight']].cov()
cov_female = df_female[['Height','Weight']].cov()
```

```python
print(corr_overall,'\n', corr_male,'\n', corr_female,'\n', cov_overall,'\n' , cov_male,'\n', cov_female,'\n')
```

## Correlation Values - Scatter Plot

```python
fig, axes = plt.subplots(1,3,figsize=(18,6))
axes[0].scatter(df['Height'],df['Weight'],color='lightgreen',alpha=0.5)
axes[0].set_title(f'Overall Correlation: {corr_overall.loc["Height","Weight"]:.2f}')
axes[0].set_xlabel('Height')
axes[0].set_ylabel('Weight')
axes[0].grid(True)

axes[1].scatter(df_male['Height'],df_male['Weight'],color='skyblue',alpha=0.5)
axes[1].set_title(f'Male Correlation: {corr_male.loc["Height","Weight"]:.2f}')
axes[1].set_xlabel('Height')
axes[1].set_ylabel('Weight')
axes[1].grid(True)

axes[2].scatter(df_female['Height'],df_female['Weight'],color='pink',alpha=0.5)
axes[2].set_title(f'Female Correlation: {corr_female.loc["Height","Weight"]:.2f}')
axes[2].set_xlabel('Height')
axes[2].set_ylabel('Weight')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

## Covariance Values - Scatter Plot

```python
fig, axes = plt.subplots(1,3,figsize=(18,6))
axes[0].scatter(df['Height'],df['Weight'],color='violet',alpha=0.5)
axes[0].set_title(f'Overall Covariance: {cov_overall.loc["Height","Weight"]:.2f}')
axes[0].set_xlabel('Height')
axes[0].set_ylabel('Weight')
axes[0].grid(True)

axes[1].scatter(df_male['Height'],df_male['Weight'],color='skyblue',alpha=0.5)
axes[1].set_title(f'Male Covariance: {cov_male.loc["Height","Weight"]:.2f}')
axes[1].set_xlabel('Height')
axes[1].set_ylabel('Weight')
axes[1].grid(True)

axes[2].scatter(df_female['Height'],df_female['Weight'],color='lightpink',alpha=0.5)
axes[2].set_title(f'Female Covariance: {cov_female.loc["Height","Weight"]:.2f}')
axes[2].set_xlabel('Height')
axes[2].set_ylabel('Weight')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

## Correlation - Heatmap

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.heatmap(corr_overall,annot=True, fmt=".2f",cmap="Purples")
plt.title('Overall Correlation Heatmap')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
